```python
from PIL import Image, ImageDraw, ImageFont
from math import floor
import matplotlib.pyplot as plt 
font_name = "Verdana.ttf"
```

```python
def shows(ims):
    f = plt.figure(figsize=(10,10))
    for i,im in enumerate(ims):
        f.add_subplot(1,len(ims),i + 1)
        plt.imshow(im)
```

```python
def pixelise(l, scale=8., cothres=128):
    w=80
    h=100
    
    t = Image.new("L", (80,100), (200,))
    d = ImageDraw.Draw(t)
    d.text((0,-20), l, font=ImageFont.truetype(font_name, size=100), fill=(0,))
    
    s = t.copy()
    s = s.resize((floor(w/scale), floor(h/scale)))
    p = s.point(lambda p: 255 if p > cothres else 0)
    shows([t,s,p])
```

```python
pixelise("R", 8., 128)
```

```python
pixelise("R", 12., 128)
```
